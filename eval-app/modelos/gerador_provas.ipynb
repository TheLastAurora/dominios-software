```python
from PIL import Image, ImageDraw, ImageFont
import random
import os
import pandas as pd
import time
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:123@localhost:5432/informacoes_concursos").connect()
font = ImageFont.truetype("consolab.ttf", 38)
```

```python
df = pd.read_sql_table('generate', engine)
```

```python
# TODO: Adicionar concursos e numero de vagas


def gen_test(ini, qnt, tipo):
    model40 = Image.open('dominios-dev\\modelos\\modelo40.jpg')
    model80 = Image.open('dominios-dev\\modelos\\modelo80.jpg')
    if (tipo == 80):
        modelo = model80
    else:
        tipo = 40
        modelo = model40
    for row in range(ini, ini+qnt):
        aux = modelo.copy()
        draw = ImageDraw.Draw(aux)
        draw.text((114, 192), df.iloc[row]['nome'], (0, 0, 0), font)
        draw.text((1148, 192), df.iloc[row]['cpf'], (0, 0, 0), font)
        draw.text((1555, 192), str(df.iloc[row]['id']), (0, 0, 0), font)
        draw.text((1665, 345), str(df.iloc[row]['num_concurso']), (0, 0, 0), font)
        draw.text((108, 345), str(df.iloc[row]['cargo']).upper(), (0, 0, 0), font)
        draw.text((108, 490), "RUA NASCIMENTO SILVA 107", (0, 0, 0), font)
        if (tipo == 80):
            x, y = random.choice([(2141, 1145), (2244, 1145), (2347, 1145)])
            draw.ellipse((x-30, y-30, x+30, y+30), fill=(0, 0, 0, 0))
            for i in [351, 890, 1427, 1965]:
                ver = [x for x in range(1431, 3105, 88)]
                hor = [x for x in range(i, (4*88)+i, 88)]
                for j in range(20):
                    x, y = random.choice(hor), ver[j]
                    draw.ellipse((x-30, y-30, x+30, y+30), fill=(0, 0, 0, 0))
        else:
            x, y = random.choice([(2192, 1145), (2295, 1145)])
            draw.ellipse((x-30, y-30, x+30, y+30), fill=(0, 0, 0, 0))
            for i in [783, 1445]:
                ver = [x for x in range(1432, 3105, 88)]
                hor = [x for x in range(i, (5*88)+i, 88)]
                for j in range(20):
                    x, y = random.choice(hor), ver[j]
                    draw.ellipse((x-30, y-30, x+30, y+30), fill=(0, 0, 0, 0))
        aux.save(
            f"RESPOSTAS\\{df.iloc[row]['id']}.jpg", optimize=True, quality=70)


gen_test(0, 8000, 80)
gen_test(8000, 9999, 40)
```
